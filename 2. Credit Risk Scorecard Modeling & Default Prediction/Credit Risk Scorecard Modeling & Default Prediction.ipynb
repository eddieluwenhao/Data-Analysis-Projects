# 特征筛选：计算每个feature的IV值


```python
import numpy as np

# 使用IV值衡量自变量的预测能力，筛选IV值>0.1的特征字段
def cal_IV(df, bin_feature, target): 
    lst = []
    # 找到feature中不同的分箱bin
    for i in range(df[bin_feature].nunique()):
        bin = list(df[bin_feature].unique())[i]
        
        # 统计feature中的一个bin的总数，以及这个bin中target=1的个数
        total = df[df[bin_feature] == bin].count()[bin_feature] # 总数
        bad = df[(df[bin_feature] == bin) & (df[target] == 1)].count()[bin_feature] # target=1的个数
        lst.append([bin_feature, bin, total, bad])
    
    # 计算'Margin_Bad'，'Margin_Good'，'WOE'，'IV'字段
    cols = ['Bin_Feature','Bin', 'Total', 'Bad']
    data = pd.DataFrame(lst, columns=cols) 
    data = data[data['Bad'] > 0]
    data['Margin_Bad'] = data['Bad'] / data['Bad'].sum()
    data['Margin_Good'] = (data['Total']-data['Bad']) / (data['Total'].sum() - data['Bad'].sum())
    data['WOE'] = np.log(data['Margin_Bad'] / data['Margin_Good'])
    data['IV'] = (data['Margin_Bad'] - data['Margin_Good']) * data['WOE']
    return data['IV'].sum()

cal_IV(df_train, 'bin_age', 'SeriousDlqin2yrs')
```

# 对于筛选出来的特征，计算每个bin的WOE值

```python
# 计算字段的WOE特征值
def cal_WOE(df, features, target):
    for feature in features:
        # sum 违约个数, count 总数
        df_woe = df.groupby(feature).agg({target: ['sum', 'count']})
        df_woe.columns = list(map(''.join, df_woe.columns.values))
        df_woe = df_woe.reset_index()
        df_woe.rename(columns={target+'sum': 'bad', target+'count': 'total'}, inplace=True)
        df_woe['good'] = df_woe['total'] - df_woe['bad']
        df_woe['Margin_Bad'] = df_woe['bad'] / df_woe['bad'].sum()
        df_woe['Margin_Good'] = df_woe['good'] / df_woe['good'].sum()
        # 为了避免分母为0，使用np.log1p
        df_woe['woe'] = np.log1p(df_woe['Margin_Bad'] / df_woe['Margin_Good'])
        df_woe.columns = [c if c==feature else c+'_'+feature for c in df_woe.columns]
        # 拼接
        df = df.merge(df_woe, on=feature, how='left')
    return df
    
cal_WOE(df_train, ['bin_age'], 'SeriousDlqin2yrs')
```

# 使用逻辑回归Logistic Regression建模

```python
from sklearn.model_selection import train_test_split, RepeatedStratifiedKFold, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, accuracy_score, roc_auc_score, auc

# 训练集切分
X = df_woe[woe_cols]
y = df_woe['SeriousDlqin2yrs']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2021, 
                                                    stratify = y) # 使得train和test的好坏分布一致
print('训练集违约率：', y_train.mean())
print('测试集违约率：', y_test.mean())
```

# 评分卡模型转换
* 设p为客户违约的概率，那么正常的概率为1-p，<br/>
Odds = p/(1-p) <br/>
* 评分卡的分值计算，可以通过 分值表示为比率对数的线性表达式来定义，即 <br/>
Score = A - B * ln(Odds) <br/>

1) A 为基准分，通常业内的基准分为500/600/650 <br/>
2) PDO( point of double odds )，即比率翻倍的分数 比如，odds翻倍时，分值减少50，即设PDO=50 <br/>
求得 B = PDO/ln2 <br/>

```python
A = 650 # 基础分
B = 72.13 # 刻度 = 50/math.log(2)

def generate_scorecard(model_coef, df_rule, features, B):
    lst = []
    cols = ['Feature', 'Binning', 'Score']
    # 模型系数
    coef = model_coef[0]
    for i in range(len(features)):
        feature = features[i]
        # 筛选feature的WOE规则
        df = df_rule[df_rule['feature'] == feature]
        for index, row in df.iterrows():
            # 逐行计算Feature, Binning, Score的值
            lst.append([feature, row['bin'], int(-row['woe']*coef[i]*B)])
    data = pd.DataFrame(lst, columns=cols)
    return data

# 生成评分卡
score_card = generate_scorecard(model.coef_, df_bin_to_woe, selected_features, B)
score_card
```
